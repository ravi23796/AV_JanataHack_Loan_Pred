```python
#Importing Libraries
#--------- ---------
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
```

```python
#Load the Dataset
#----------------
train=pd.read_csv('train_ctrUa4K.csv')
test=pd.read_csv('test_lAUu6dG.csv')
train_orig=train
test_orig=test
```

```python
#Analysis
#--------
print("Train Data Datatypes")
print("----- ---- ---------")
print(train.dtypes)
print()
print("Train Data Shape")
print("----- ---- -----")
print(train.shape)
print()
print("Target Variable")
print("------ --------")
train['Loan_Status'].value_counts(normalize=True).plot.bar()
plt.show()

print("Categorical Variables")
print("----------- ---------")
plt.figure(1)
plt.subplot(221)
train['Gender'].value_counts(normalize=True).plot.bar(figsize=(20,10),title="Gender")

plt.subplot(222)
train['Married'].value_counts(normalize=True).plot.bar(title="Married")

plt.subplot(223)
train['Self_Employed'].value_counts(normalize=True).plot.bar(title="Self Employed")

plt.subplot(224) 
train['Credit_History'].value_counts(normalize=True).plot.bar(title= 'Credit_History') 
plt.show()

print("Ordinal Variables")
print("------- ---------")
plt.figure(1) 
plt.subplot(131) 
train['Dependents'].value_counts(normalize=True).plot.bar(figsize=(24,6), title= 'Dependents') 

plt.subplot(132) 
train['Education'].value_counts(normalize=True).plot.bar(title= 'Education') 

plt.subplot(133) 
train['Property_Area'].value_counts(normalize=True).plot.bar(title= 'Property_Area') 

plt.show()
```

```python
plt.figure(1)

plt.subplot(121)
sns.distplot(train['ApplicantIncome'])

plt.subplot(122)
train['ApplicantIncome'].plot.box(figsize=(15,5))
```

```python
plt.figure(1)

plt.subplot(121)
sns.distplot(train['LoanAmount'])

plt.subplot(122)
train['LoanAmount'].plot.box(figsize=(15,5))

plt.show()
```

```python
Gender=pd.crosstab(train['Gender'],train['Loan_Status']) 
Gender.div(Gender.sum(0).astype(float), axis=0).plot(kind="bar", stacked=True, figsize=(4,4))
```

```python
Gender=pd.crosstab(train['Gender'],train['Loan_Status'])
Gender.div(Gender.sum(1).astype(float),axis=0).plot.bar(stacked=True,figsize=(2,2))

Married=pd.crosstab(train['Married'],train['Loan_Status'])
Married.div(Married.sum(1).astype(float),axis=0).plot.bar(stacked=True,figsize=(2,2))

Dependents=pd.crosstab(train['Dependents'],train['Loan_Status'])
Dependents.div(Dependents.sum(1).astype(float),axis=0).plot.bar(stacked=True,figsize=(2,2))

Education=pd.crosstab(train['Education'],train['Loan_Status'])
Education.div(Education.sum(1).astype(float),axis=0).plot.bar(stacked=True,figsize=(2,2))

Self_Employed=pd.crosstab(train['Self_Employed'],train['Loan_Status'])
Self_Employed.div(Self_Employed.sum(1).astype(float),axis=0).plot.bar(stacked=True,figsize=(2,2))

Credit_History=pd.crosstab(train['Credit_History'],train['Loan_Status'])
Credit_History.div(Credit_History.sum(1).astype(float),axis=0).plot.bar(stacked=True,figsize=(2,2))

Property_Area=pd.crosstab(train['Property_Area'],train['Loan_Status'])
Property_Area.div(Property_Area.sum(1).astype(float),axis=0).plot.bar(stacked=True,figsize=(2,2))
```
